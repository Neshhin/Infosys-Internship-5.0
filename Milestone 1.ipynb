```python
# Loading dataset
df = pd.read_csv('household_power_consumption.txt',sep=";")
```

```python
# Check the shape of the dataset (rows, columns)
print("Dataset Shape:", df.shape)
```

```python
# Display column names with their respective datatypes
print("Column Data Types:")
print(df.dtypes)
```

```python
# Check for missing values in each column
print("Missing Values Per Column:")
print(df.isnull().sum())
```

```python
# Drop or fill missing values as needed
# For demonstration, let's use `dropna()` to remove rows with missing values.
df_cleaned = df.dropna()
```

```python
# Validate if there are no missing values after cleaning
print("Missing Values After Cleaning:")
print(df_cleaned.isnull().sum())
```

```python
# Display unique values for each column to gain insights
for col in df.columns:
    unique_values = df[col].unique()
    print(f"Unique values in '{col}': {unique_values}")
```
