# Importing Matplotlib and Seaborn
# **Matplotlib**
For Matplotlib, we don’t need to import the entire library. Instead, we only import its sub-module pyplot, which simplifies plotting tasks.
We’ll use the alias name plt for convenience.

# What is pyplot?
pyplot is a sub-module of Matplotlib designed for data visualization.
Think of it as a high-level API that makes plotting easier and more intuitive.
Data scientists primarily use pyplot for their visualizations unless they aim to create something entirely custom.

# Seaborn
For Seaborn, we import the entire library using the alias sns.
# What is Seaborn?
Seaborn is another visualization library that builds on Matplotlib for its plotting backend.

It simplifies the visualization process by:
Using aesthetic themes to make plots more visually appealing.
Automating many aspects of data visualization, reducing the number of code lines needed.

Major Differences Between Matplotlib and Seaborn
Underlying Framework:

Matplotlib provides the foundation for plotting, allowing you to build basic visualizations with high customizability.
Seaborn enhances Matplotlib, making use of Pandas data structures and simplifying complex visualizations.
Themes and Aesthetics:

Seaborn includes built-in themes and style customizations for polished visuals, while Matplotlib requires more manual customization.
Code Efficiency:

Seaborn minimizes the code required by automating many tasks in the background.
Matplotlib requires explicit coding for similar results.

# Components of a Matplotlib Plot
# 1. Figure
The Figure is the overall window or page where all visual elements are drawn.

Key Points:

You can create multiple independent Figures in Jupyter Notebook.

If running code in a terminal, each Figure opens in a separate window.

# 2. Axes

The Axes is the specific area within the Figure where the data is plotted.
Multiple Axes can be added to a single Figure to represent different plots.

# 3. Axis
Refers to the x-axis and y-axis of the plot.

# 4. x-label and y-label
The labels for the x-axis and y-axis, providing context for the data being plotted.

# 5. Ticks

**Major Ticks:**

Divide the axis into major units.

Displayed by default when plotting.

**Minor Ticks:**

Further subdivide the major tick units.

Hidden by default but can be enabled as needed.

# 6. Title and Subtitle

Title: Refers to the title for each individual plot (Axes).

Subtitle: A common title for all plots within a Figure.

# 7. Legend

Explains the elements in the plot.

Example: Blue and green curves representing different datasets.



These are the primary components of a Matplotlib plot. Understanding them helps in creating and customizing visualizations effectively.

# **Visualizing Categorical Data with a Bar Chart or Count Plot**

To visualize categorical data, we often use a Bar Chart or a Count Plot, where:

**X-axis:** Represents the categories.

**Y-axis:** Represents the corresponding frequencies of those categories.

## **Bar Chart**
A Bar Chart is a great choice for displaying categorical data.
The data is grouped (or "binned") into categories, and the height of each bar reflects the frequency of that category.

**How to Create a Bar Chart?**

Use plt.bar() from Matplotlib to create a bar chart.

Example: Visualizing platform Frequencies
Step-by-Step:
Group the data into categories (e.g., count each unique platform).
Plot these categories on the X-axis and their frequencies on the Y-axis.

```python
plt.plot(count)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Platform')
plt.ylabel('No. of platforms')
plt.title('paltform & count', fontsize=15)

```

# **Pie Chart**

**When to Use a Pie Chart?**

Use a Pie Chart when you want to visualize the proportion or distribution of categories relative to the whole.

Example: Comparing the proportion of sales across different regions.

**Key Features of a Pie Chart:**

Each slice represents a category, with its size proportional to its share in the total.

Best used for a small number of categories (e.g., < 7) to avoid clutter.

Example: Visualizing Proportions of Sales Across Regions

Step-by-Step:

Calculate the total sales for all regions.

Determine the proportion (or percentage) of sales for each region.

Use plt.pie() to create the Pie Chart.

```python
sales_data = data[['JP_Sales', 'EU_Sales', 'NA_Sales', 'Other_Sales']]
region_sales = sales_data.sum()

plt.pie(region_sales,
       labels=region_sales.index,
       startangle=90,
       explode=(0.01,0.01,0.01,0.1),
       autopct = '%.2f%%') # label the wedges with their numeric value # most prefferd start is 90 degree

plt.title('Total Sales')
plt.show()
```

# **Histogram**

**What is a Histogram?**

A Histogram is a visualization tool for showing the distribution of numerical data.

The data is divided into intervals (called bins), and the frequency of data points within each bin is represented as a bar.

**Key Features of a Histogram:**

X-axis: Represents the range of data (bins).

Y-axis: Represents the frequency (count of data points within each bin).

Useful for identifying patterns like skewness, spread, peaks, or outliers in the data.

**When to Use a Histogram?**

To analyze the distribution of a single numerical variable.

Examples:

Distribution of player scores in a video game.

Examining the age distribution of a customer base.

Identifying patterns in sales revenue.

Example: Visualizing the Distribution of Scores

**sns.kdeplot and Its Purpose**

**What Does sns.kdeplot() Do?**

sns.kdeplot() is a function in Seaborn used to create a Kernel Density Estimate (KDE) plot.

A KDE plot is a smoothed, continuous curve that represents the probability density of a numerical variable, offering an alternative to histograms.

**Key Features of a KDE Plot:**

Visualization of Distribution:

Shows the shape (e.g., unimodal, bimodal) and spread of the data.

Smoothing:

Uses a kernel function to smooth the data distribution.

No Binning:

Unlike histograms, KDE plots don’t divide the data into bins, providing a more fluid representation.

```python
sns.kdeplot(data['Year'])
plt.show()
```

# **Box Plot**

**What is a Box Plot?**

A Box Plot (or Box-and-Whisker Plot) is a visualization used to summarize the distribution of numerical data through:

Median: The central value.

Quartiles: Divides the data into four equal parts.

Whiskers: Indicates the spread of the data outside the quartiles.

Outliers: Data points that lie significantly outside the range.

**Key Features of a Box Plot:**

Provides a visual summary of:

Central Tendency: Using the median line.

Variability: Through the interquartile range (IQR).

Symmetry or Skewness: Based on the position of the median and whiskers.

Outliers: Highlighting data points outside the whisker range.

**When to Use a Box Plot?**

To compare distributions across categories.

To detect skewness, variability, and outliers in the data.

Example: Comparing the scores of students across different subjects or sales revenue across regions.


```python
plt.boxplot(data['Other_Sales'])
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.boxplot(y = data["Other_Sales"])
plt.yticks(fontsize=20)
plt.ylabel('Other Sales (in million dollars)', fontsize=20)
plt.xticks(rotation=60)
plt.title('Other Sales of video games', fontsize=20)
```

# **Scatter Plot**

**What is a Scatter Plot?**

A Scatter Plot is a type of visualization that uses points to display the relationship between two numerical variables.

Each point represents a single observation in the dataset, with:

X-axis: One variable.

Y-axis: Another variable.

**Key Features of a Scatter Plot:**

Correlation:

Shows if the variables have a positive, negative, or no correlation.

Trend Identification:

Helps identify linear, non-linear, or clustered trends.

Outliers:

Points that lie far away from the general data pattern are easily noticeable.

**When to Use a Scatter Plot?**

To visualize the relationship between two continuous numerical variables.

Examples:

Height vs. Weight of individuals.

Revenue vs. Advertising spend.

Age vs. Salary.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='EU_Sales', y='Other_Sales', hue='Genre', data=data) # the variable should be both numeric
plt.title('EU Sales vs. Other Sales by Genre')
plt.xlabel('EU Sales (in millions)')
plt.ylabel('Other Sales (in millions)')
plt.show()
```

# **Heatmap**

**What is a Heatmap?**

A Heatmap is a visualization that uses color intensity to represent data values in a grid.

It's commonly used for:

Showing correlations between variables.

Representing numerical data in a matrix-like format (e.g., confusion matrices or pivot tables).

**Key Features of a Heatmap:**

Color Gradient:

Darker or lighter colors indicate the intensity of the values.

Compact Representation:

Allows you to quickly grasp patterns and relationships in large datasets.

Grid-Like Layout:

Rows and columns are clearly labeled, making it ideal for comparing data

categories.

**When to Use a Heatmap?**

Visualizing Relationships:

Example: Correlation matrix for multiple numerical variables.

Summarizing Aggregated Data:

Example: Sales performance by region and month.

Machine Learning Metrics:

Example: Visualizing confusion matrices.


```python
# Group by Publisher and Genre, then sum the sales for each combination
publisher_genre_sales = data.groupby(['Publisher', 'Genre'])['Global_Sales'].sum().unstack()

# Get the top 10 publishers based on total global sales
top_10_publishers = data.groupby('Publisher')['Global_Sales'].sum().nlargest(10).index

# Filter the sales data to include only the top 10 publishers
publisher_genre_sales = publisher_genre_sales.loc[top_10_publishers]

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(publisher_genre_sales, annot=True, fmt=".2f", cmap="viridis")
plt.title("Heatmap of Global Sales by Top 10 Publishers and Genre")
plt.xlabel("Genre")
plt.ylabel("Publisher")
plt.xticks(rotation=60)
plt.show()
```

# **1. Convert Date and Time from String to Date (Year, Month, Day) & Time (Hours, Minutes)**
We'll assume the dataset has a column with date and time in string format, like "YYYY-MM-DD HH:MM:SS". Here's how to convert it:

```python


import pandas as pd


```

```python
# Sample DataFrame
data = {'datetime': ['2024-11-12 06:30:00', '2024-11-12 19:45:00']}
df = pd.DataFrame(data)
```

```python
# Convert the datetime column to pandas datetime format
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
# Extract Date (Year, Month, Day) and Time (Hour, Minute)
df['date'] = df['datetime'].dt.date
df['time'] = df['datetime'].dt.strftime('%H:%M')
```

```python
print(df)
```

# **2. Find Whether the Day is a Holiday or Not (0 = Not Holiday, 1 = Holiday)**
You can create a list of holidays or use a public holiday API to check if the day is a holiday.

```python
# List of sample holidays (for demonstration purposes)
holidays = ['2024-11-12', '2024-12-25']

# Check if the date is a holiday
df['holiday'] = df['date'].apply(lambda x: 1 if str(x) in holidays else 0)

print(df)
```

# **3. Find If There Is Sunlight During the Time Given in the Dataset (0600-1800 hours — 1 for Light, 0 for No Light)**

```python
# Define the sunlight hours range (0600 - 1800)
def check_sunlight(time):
    hour = int(time.split(':')[0])  # Get the hour from the time string
    return 1 if 6 <= hour < 18 else 0

# Apply the function to check sunlight
df['sunlight'] = df['time'].apply(check_sunlight)

print(df)
```
